# MATPLOTLIB LABELS AND TITLE

```python
# WE USE XLABEL() AND YLABEL() FOR LABELING OF X AND Y AXIS
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.plot(x,y)
plt.xlabel('Average pulse of individual')
plt.ylabel('No. of calories burned')
```

```python
# to create a title of data we use title() keyword
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.plot(x,y)
plt.xlabel('Average pulse of individual')
plt.ylabel('No. of calories burned')
plt.title('HEALTH CAMP GRAPH')
```

```python
# WE CAN USE FONTDICT TO SELECT FONT FOR OUR VISUALIZATIONS

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.plot(x,y)
plt.xlabel('Average pulse of individual',fontdict=font2)
plt.ylabel('No. of calories burned',fontdict=font2)
plt.title('HEALTH CAMP GRAPH',fontdict=font1)
```

# MATPLOTLIB ADDING GRID LINES

```python
# WE CAN ADD GRID LINES BY USING GRID FUNCTIONS
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.grid()
plt.plot(x,y)
plt.xlabel('Average pulse of individual',fontdict=font2)
plt.ylabel('No. of calories burned',fontdict=font2)
plt.title('HEALTH CAMP GRAPH',fontdict=font1)
```

```python
# only single axis line to display 
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.grid(axis='x')
plt.plot(x,y)
plt.xlabel('Average pulse of individual',fontdict=font2)
plt.ylabel('No. of calories burned',fontdict=font2)
plt.title('HEALTH CAMP GRAPH',fontdict=font1)
```

```python
#grid lines properties
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.grid(color = 'r',ls='--',lw =1.5)
plt.plot(x,y)
plt.xlabel('Average pulse of individual',fontdict=font2)
plt.ylabel('No. of calories burned',fontdict=font2)
plt.title('HEALTH CAMP GRAPH',fontdict=font1)
```

# MATPLOTLIB SCATTER

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x,y)
plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x,y)
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x,y)
plt.show()
```

```python
# for each scatter plot color can be changed by using c
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x,y,color ='r')
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x,y,color='g')
plt.show()
```

```python
#we can color each dot diffrently by passing an color np.array
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
color = np.array([0,10,20,30,35,40,45,50,60,70,80,90,100])
plt.scatter(x,y,c=color,cmap='viridis')
plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
color = np.array([0,10,20,30,35,40,45,50,60,70,80,90,100])
plt.scatter(x,y,c=color,cmap='viridis')
plt.colorbar()
plt.show()
```

```python
# we can change the size of dots by using s keyword
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
plt.scatter(x,y,s=sizes)
plt.show()
```

```python
# in this we can also use alpha for the transparency of the dot
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
plt.scatter(x,y,s=sizes,alpha=0.5)
plt.show()
```

```python
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```
